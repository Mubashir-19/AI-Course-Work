Question 1: What is the average ratings in the dataset?

```python
import pandas as pd
import matplotlib.pyplot as plt

movie_data = pd.read_csv("movies.csv")

plt.hist(movie_data['Vote_average'], bins=20, color='cyan', edgecolor='black')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Movie Ratings')
plt.grid(True)
plt.show()
```

Question 2: Which movie is the best?

Ans: Movie with Highest Rating and Highest Votes is the best. The Scatter Plot below shows this relationship.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('movies.csv')

# Calculate average rating and average votes
data['Average_rating'] = data['Vote_average']
data['Average_votes'] = data['Vote_count']
average_data = data.groupby('Name').agg({'Average_rating': 'mean', 'Average_votes': 'mean'}).reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(average_data['Average_rating'], average_data['Average_votes'], alpha=0.5)
plt.title('Average Rating vs Average Votes for Movies')
plt.xlabel('Average Rating')
plt.ylabel('Average Votes')
plt.grid(True)
plt.show()
```

Question 3: Which are the latest movies?

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("movies.csv")


df['First_air_date'] = pd.to_datetime(df['First_air_date'], dayfirst=True)

# Sort DataFrame by 'First_air_date'
df = df.sort_values(by='First_air_date')

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df['First_air_date'], df['Vote_average'], marker='o')
plt.title('Movie Ratings Over Time')
plt.xlabel('Release Date (Oldest to Newest)')
plt.ylabel('Rating')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Question 4: Is there any correlation between the vote average and the vote count?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset from CSV
df = pd.read_csv('movies.csv')

# Calculate the Pearson correlation coefficient
correlation = df['Vote_average'].corr(df['Vote_count'])

print("Pearson correlation coefficient between vote average and vote count:", correlation)

# Plot vote average vs. vote count
plt.figure(figsize=(8, 6))
plt.scatter(df['Vote_average'], df['Vote_count'], alpha=0.5)
plt.title('Vote Average vs. Vote Count')
plt.xlabel('Vote Average')
plt.ylabel('Vote Count')
plt.grid(True)
plt.show()
```

Question 5: Which movies have recieved the most votes?

```python
import pandas as pd

# Load the dataset from CSV
df = pd.read_csv('movies.csv')

# Sort the DataFrame by 'Vote_count' in descending order
df_sorted = df.sort_values(by='Vote_count', ascending=False)

# Select the top entries with the most votes
top_voted = df_sorted.head(10)  # Change 10 to the desired number of top entries

print("Top voted movies or TV shows:")
print(top_voted[['Name', 'Original_name', 'Vote_count']])
```

Question 6: Are there any trends in the popularity of movies or TV shows over time?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from CSV
df = pd.read_csv('movies.csv')

# Convert 'First_air_date' column to datetime
df['First_air_date'] = pd.to_datetime(df['First_air_date'])

# Group by year and calculate the average popularity
popularity_trends = df.groupby(df['First_air_date'].dt.year)['Popularity'].mean()

# Plot the popularity trends over time
plt.figure(figsize=(10, 6))
popularity_trends.plot(marker='o')
plt.title('Popularity Trends of Movies/TV Shows Over Time')
plt.xlabel('Year')
plt.ylabel('Average Popularity')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
